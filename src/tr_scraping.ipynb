```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from bs4 import BeautifulSoup

```

```python
# NOTE: This might not work based on your set up
# Set up Selenium WebDriver
options = Options()
options.headless = True  # Run without opening a browser window
driver = webdriver.Firefox(options=options)

# Load the page
url = "https://www.transfermarkt.com/premier-league/transfers/wettbewerb/GB1/saison_id/2022"
driver.get(url)

# Wait for JavaScript to execute
# (You may need to adjust the waiting time based on the page load speed)
driver.implicitly_wait(5)  # Wait for 10 seconds

# Get the page source after JavaScript has executed
page_source = driver.page_source

# Close the WebDriver
driver.quit()

# Parse the page source with BeautifulSoup
soup = BeautifulSoup(page_source, "html.parser")

# Now you can continue with parsing the content using BeautifulSoup
```

### Getting squad transfer sum info

Get squad transfer sum info

```python
transfer_sum_box = soup.find_all('div', class_='transfer-zusatzinfo-box')
```

```python
len(transfer_sum_box)
```

```python
transfer_sum_box[0]
```

```python
tr_soup = BeautifulSoup(str(transfer_sum_box[0]), 'html.parser')
```

```python
tr_soup
```

```python
data = {}
for span in tr_soup.find_all('span'):
  text = span.text.strip()  # Remove leading/trailing whitespace
  if text:
    key, value = text.split(':')  # Split at the colon (":")
    data[key.strip()] = value.strip()  # Remove extra whitespace from key and value
```

```python
df = pd.DataFrame(data, index=[0])  # Create DataFrame with a single row

print(df)
```
