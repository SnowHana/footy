# 1. Get Team URLs

```python
import requests
from bs4 import BeautifulSoup
```

```python
standings_url = "https://fbref.com/en/comps/9/Premier-League-Stats"
data = requests.get(standings_url)
# data.text
```

```python
soup = BeautifulSoup(data.text)
```

```python
standings_table = soup.select('table.stats_table')[0]
standings_table
```

```python
links = standings_table.find_all('a')
links = [l.get('href') for l in links]
links = [l for l in links if '/squads/' in l]
links
```

```python
team_urls = [f"https://fbref.com{l}" for l in links]
team_urls
```

## 2. Store Squads and Team
We got the team url. Access one by one and get the squad list, store them to csv file


```python
import pandas as pd
import time
```

```python
team_url = team_urls[1]
data = requests.get(team_url)
squads = pd.read_html(data.text, match="Standard Stats")[0]
# Drop multi level indexing
squads
```

```python
# Remove multilevel indexing
squads = squads.droplevel(level=0,axis=1)
squads
```
